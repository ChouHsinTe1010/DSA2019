```python
# Python3 Program to print BFS traversal 
# from a given source vertex. BFS(int s) 
# traverses vertices reachable from s. 
from collections import defaultdict 
  
# This class represents a directed graph 
# using adjacency list representation 
class Graph:
    # Constructor 
    def __init__(self): 
        # default dictionary to store graph 
        self.graph = defaultdict(list) 

    # function to add an edge to graph 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
  
    # Function to print a BFS of graph 
    def BFS(self, s):
        ans=[]
        if s in self.graph:    #起點有沒有在圖形中
            repeated=defaultdict(list)#設立一個重複值放的list
            queue = [s]#把第一個數放進queue
            repeated[s]=True#第一個值在重複值裡面
            while queue:
                vertex=queue.pop(0)#pop第一個數出去，此為BFS的規則
                ans.append(vertex)#pop之後加入ans裡
                for i in self.graph[vertex]:
                    #if repeated[i] == False
                    if not repeated[i]:   #如果沒拜訪過則執行
                        repeated[i]=True
                        queue.append(i)
        return ans          
            
        
    
    def DFS(self, s):
        ans=[]
        if s in self.graph:#起點有沒有在圖形中
            repeated=defaultdict(list)
            stack =[s]#把第一個數放進stack
            repeated[s]=True#第一個值在重複值裡面
            while stack:
                vertex = stack.pop()#pop最後一個數出去，此為DFS的規則
                ans.append(vertex)#pop之後加入ans裡
                for i in self.graph[vertex]:
                    if not repeated[i]: #如果沒拜訪過則執行
                        repeated[i]=True
                        stack.append(i)
                            
        return ans
        
g=Graph()
g.addEdge(0,1)
g.addEdge(0,2)
g.addEdge(1,2)
g.addEdge(2,0)
g.addEdge(2,3)
g.addEdge(3,3)
print(g.BFS(2))
print(g.DFS(2))
#print(g.graph)                      
```
